```python
from numpy.core.fromnumeric import mean as mn
```

```python
from datetime import datetime as dt
import matplotlib.pyplot as plt
from meteostat import Point,Daily
import numpy as np
```

```python
start=dt(2023,3,1)
end=dt(2023,3,31)
Kolkata=Point(22.5726,88.3639,10)
Barrackpur=Point(22.766,88.3516,10)
```

```python
#Get daily data for 2023
daily_data=Daily(Kolkata,start,end)
my_data=daily_data.fetch()
num_data=my_data['wspd'].to_numpy()
print("city mean= ",np.mean(num_data))
```

```python
daily_data2=Daily(Barrackpur,start,end)
my_data2=daily_data2.fetch()
num_data2=my_data2['wspd'].to_numpy()
print("suburbs mean= ",np.mean(num_data2))
```

```python
#plot line chart
x=np.linspace(1,31,31)
plt.plot(x,my_data['wspd'],label="kol wspd")
plt.plot(x,my_data2['wspd'],label="bkp wspd")
plt.legend()
plt.title("kol wspd v/s bkp wspd",color="red")
plt.xlabel("Days",color="red")
plt.ylabel("Wind speed (km/h)",color="red")
plt.grid(True)
plt.show()
```
